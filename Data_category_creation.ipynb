```python
url = "https://drive.google.com/file/d/1d65Rm9fxxYl2XHFWUMABZYFWCG-tf0fj/view?usp=sharing" # brands.csv
path = "https://drive.google.com/uc?export=download&id="+url.split("/")[-2]
brands = pd.read_csv(path)
```

```python
product_category_df=product_category_df.merge(brands, left_on="short_sku", right_on='short', how="left")
```

```python
TYPE_MAP = {
    "113281716": ("Phone", "iPhone 8"),
    "200100001": ("Monitor", "Apple Studio Display"),
    "113271716": ("Phone", "iPhone X"),
    "106431714": ("Tablet", "iPad Pro"),
    "10230": ("Laptop", "Laptop Accessory"),
    "101781405": ("Tablet", "Tablet Accessory"),
    "10142": ("Accessory", "Battery"),
    "13855401": ("Keyboard", "Keyboard"),
    "11935397": ("Hard Drive / SSD", "HDD/SSD"),
    "12655397": ("Hard Drive / SSD", "HDD/SSD"),
    "12215397": ("Hard Drive / SSD", "HDD/SSD"),
    "1364": ("Computer Components", "RAM"),
    "54025401": ("Keyboard", "Keyboard"),
    "11865403": ("Accessory", "Case")
}
```

```python
type(TYPE_MAP)
```

```python
phone_brands = {"apple", "samsung", "xiaomi", "huawei", "oppo"}
laptop_brands = {"lenovo", "microsoft", "asus", "dell", "hp","apple"}
headphone_brands = {"beats", "bose", "jbl", "sony"}
wearable_brands = {"band&strap", "fitbit", "garmin"}
monitor_brands = {"lg", "benq", "dell", "apple","asus", "samsung"}

re_iphone= re.compile(r"\biphone\s*(\d{1,2})(?:\s*(pro|pro max| plus|max))?\b", re.I)
re_ipad= re.compile(r"\bipad\b(?:\s*(pro|air|mini))?\s*(\d{1,2})?\b", re.I)
re_mac= re.compile(r"\bmacbook\b(?:\s*(air|pro))?\b", re.I)
re_watch  = re.compile(r"\bapple\s*watch\b(?:\s*([su]\d+|ultra|se))?", re.I)
re_airpods= re.compile(r"\bairpods\b(?:\s*(pro|max))?", re.I)
re_galaxy = re.compile(r"\bgalaxy\s*([a-z]?\d{1,2}|z\s*fold|z\s*flip|s\d{1,2}|a\d{1,2})\b", re.I)
re_monitor= re.compile(r"\b(studio\s*display|pro\s*display\s*xdr|monitor|display|screen)\b", re.I)
re_laptop = re.compile(r"\b(laptop|notebook|ultrabook)\b", re.I)
re_tablet = re.compile(r"\b(tablet)\b", re.I)
re_phone  = re.compile(r"\b(smart?phone|handy|telefon)\b", re.I)
storage_drive = re.compile(
    r"\b(hard\s*drive|external\s*hdd|external\s*ssd|portable\s*drive|disk\s*drive|harddisk|ssd\s*drive|usb\s*drive|flash\s*drive|memory\s*stick)\b",
    re.I,
)


acc_case = re.compile(r"\b(case|cover)\b", re.I)
acc_cable     = re.compile(r"\b(cable|charger|charging|adapter|dock|usb|lightning|mag?safe|power\s*bank)\b", re.I)
acc_protector = re.compile(r"\b(screen\s*protector|tempered\s*glass|anti-scratch|schutzfolie)\b", re.I)
acc_strap     = re.compile(r"\b(strap|band|bracelet|kayÄ±ÅŸ|armband)\b", re.I)
acc_audio     = re.compile(r"\b(earbuds?|earphones?|headset)\b", re.I)
acc_keyboard  = re.compile(r"\b(keyboard|klavye|key\s*pad|magic\s*keyboard|smart\s*keyboard)\b", re.I)
acc_ram       = re.compile(r"\b(ram|memory)\b", re.I)

```

```python
def _text(*vals):
    return " ".join([str(v) for v in vals if pd.notna(v)]).lower()
```

```python
def categorize_row(row, type_map = TYPE_MAP):
  brand = str(row.get("long", "") or "").strip().lower()
  name = str(row.get("name", "") or "")
  desc = str(row.get("desc", "") or "")
  typ = str(row.get("type", "") or "")

  text = _text(name, desc)

  if typ in type_map:
    return type_map[typ]

  m=re_iphone.search(text)
  if m:
    model=m.group(1)
    suffix = (m.group(2) or "").upper().replace("PRO MAX", "Pro Max").title()
    detailed = f"iPhone {model}" + (f"{suffix}" if suffix else "")
    return ("Phone", detailed)

  m = re_ipad.search(text)
  if m:
      family = m.group(1)
      gen    = m.group(2)
      detail = "iPad"
      if family: detail += f" {family.title()}"
      if gen:    detail += f" {gen}"
      return ("Tablet", detail)

  m = re_mac.search(text)
  if m:
      variant = m.group(1)
      detail = "MacBook" + (f" {variant.title()}" if variant else "")
      return ("Laptop", detail)

  m = re_mac.search(text)
  if m:
      variant = m.group(1)
      detail = "MacBook" + (f" {variant.title()}" if variant else "")
      return ("Laptop", detail)

  # Apple Watch
  m = re_watch.search(text)
  if m:
      variant = m.group(1)
      detail = "Apple Watch" + (f" {variant.upper()}" if variant else "")
      return ("Wearable", detail)

  # AirPods
  m = re_airpods.search(text)
  if m:
      variant = m.group(1)
      detail = "AirPods" + (f" {variant.title()}" if variant else "")
      return ("Headphones", detail)

  # Samsung Galaxy
  m = re_galaxy.search(text)
  if m or (brand == "samsung" and re_phone.search(text)):
      detail = "Samsung Galaxy"
      if m:
          model = m.group(1).upper().replace(" ", "")
          detail += f" {model}"
      return ("Phone", detail)

  if re_monitor.search(text) or brand in monitor_brands and ("display" in text or "monitor" in text):

      if "studio display" in text:
          return ("Monitor", "Apple Studio Display") if brand == "apple" else ("Monitor", "Studio Display")
      if "pro display xdr" in text:
          return ("Monitor", "Apple Pro Display XDR") if brand == "apple" else ("Monitor", "Pro Display XDR")
      return ("Monitor", f"{row.get('brand_name','').strip()} Monitor".strip())

  if re_laptop.search(text) or (brand in laptop_brands and ("book" in text or "notebook" in text or "laptop" in text)):
      return ("Laptop", f"{row.get('brand_name','').strip()} Laptop".strip())

  if re_tablet.search(text):
      return ("Tablet", f"{row.get('brand_name','').strip()} Tablet".strip())

  if re_phone.search(text) or brand in phone_brands and any(k in text for k in ["phone","iphone","galaxy","android","telefon","handy"]):
      return ("Phone", f"{row.get('brand_name','').strip()} Phone".strip())

  if acc_case.search(text):
      return ("Accessory", "Phone Case")
  if acc_protector.search(text):
      return ("Accessory", "Screen Protector")
  if acc_strap.search(text):
      if "watch" in text or "apple watch" in text:
          return ("Accessory", "Watch Strap/Band")
      return ("Accessory", "Strap/Band")
  if acc_cable.search(text):
      if "lightning" in text:
          return ("Accessory", "Lightning Cable/Charger")
      if "usb-c" in text or "type-c" in text:
          return ("Accessory", "USB-C Cable/Charger")
      if "magsafe" in text or "mag safe" in text:
          return ("Accessory", "MagSafe Charger")
      return ("Accessory", "Charger/Cable")
  if acc_keyboard.search(text):
    if "magic keyboard" in text:
        return ("Apple Keyboard", "Magic Keyboard")
    if "smart keyboard" in text:
        return ("Computer Components", "Smart Keyboard")
    return ("Computer Components", "Keyboard")
  if acc_ram.search(text):
    return ("Computer Components", "RAM")

  if storage_drive.search(text):
    if "ssd" in text:
        return ("Hard Drive / SSD", "Solid State Drive (SSD)")
    elif "hdd" in text or "hard" in text or "disk" in text:
        return ("Hard Drive / SSD", "Hard Disk Drive (HDD)")
    elif "usb" in text or "flash" in text:
        return ("Hard Drive / SSD", "USB Flash Drive")
    else:
        return ("Hard Drive / SSD", "External Drive")


  if acc_audio.search(text) or brand in headphone_brands:
      return ("Headphones", f"{row.get('brand_name','').strip()} Headphones".strip())

  if brand in wearable_brands or "watch" in text:
      return ("Wearable", f"{row.get('brand_name','').strip()} Wearable".strip())
  if brand:
      return ("Accessory", f"{row.get('brand_name','').strip()} Accessory".strip())


  return ("Other", "Other")
```

```python
cats = product_category_df.apply(lambda row: pd.Series(categorize_row(row)), axis=1)
cats.columns = ["KS_general_category", "KS_detailed_category"]

```

```python
def guess_general_category(brand):
    if pd.isna(brand):
        return "Other"
    brand = brand.lower()
    if brand in ["apple", "samsung", "xiaomi", "huawei"]:
        return "Phone"
    elif brand in ["digital", "western digital","pack", "seagate","lacie", "crucial","sandisk","owc", "lexar","g-technology", "glyph atom","kingston","trascend"]:
        return "Hard Drive/SSD"
    elif brand in ["lenovo", "microsoft", "asus"]:
        return "Laptop"
    elif brand in ["beats", "bose", "jbl"]:
        return "Headphones"
    elif brand in ["band&strap", "fitbit", "garmin"]:
        return "Wearable"
    elif brand in ["lg", "benq", "dell", "apple"]:
        return "Monitor"
    else:
        return "Accessory"
```
